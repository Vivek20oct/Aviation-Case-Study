```python
df['fatality_level'] = df['Injury.Severity'].str.extract(r'(\d+)', expand=False)
df['fatality_level'] = pd.to_numeric(df['fatality_level'], errors='coerce')
```

```python
def numeric_missing_values(df,colmn):
    for col in colmn:
        df[col].fillna(0,inplace=True)
    return df
```

```python
columns=['Total.Uninjured' , 'Total.Minor.Injuries' ,'Total.Serious.Injuries' , 'Total.Fatal.Injuries' , 'fatality_level']
```

```python
df=numeric_missing_values(df,columns)
```

```python
df['Number.of.Engines'].fillna(df['Number.of.Engines'].median(),inplace=True)
```

```python
df['Event.Date']=pd.to_datetime(df['Event.Date'])
```

```python
df['Publication.Date']=pd.to_datetime(df['Publication.Date'])
```

```python
df['Publication.Date'].fillna(pd.Timestamp('1800-01-01'),inplace=True)
```

```python
object_columns = df.select_dtypes(include=['object']).columns
df[object_columns] = df[object_columns].fillna('NoData')

# 2. Replace 'N/A' values in object columns with "NoData"
df[object_columns] = df[object_columns].replace('N/A', 'NoData')
```

# 3. Data Preprocessing:

```python
def get_IQR(df,colmn):
    for col in colmn:
        q1=df[col].quantile(0.25)
        q3=df[col].quantile(0.75)
        IQR=q3-q1
        lower_bound=q1-1.5*IQR
        upper_bound=q3+1.5*IQR
        print(f'Outlier of {col} , Q1={q1} , Q3={q3} , IQR={IQR} , Lower_bound={lower_bound} , Upper_bound={upper_bound}')
```

```python
columns=['Number.of.Engines','Total.Uninjured' , 'Total.Minor.Injuries' ,'Total.Serious.Injuries' , 'Total.Fatal.Injuries']
```

```python
new_df=df[(df['Total.Uninjured']!=0) | (df['Total.Minor.Injuries']!=0) | (df['Total.Serious.Injuries']!=0) | (df['Total.Fatal.Injuries']!=0)| ['fatality_level'] != 0]
```

Q7. Impact of Flight Phase on Severity: How does the broad phase of flight impact injury severity or aircraft damage? Use boxplots or violin plots to show injury severity across flight phases.

```python
columns_to_plot = ['Total.Serious.Injuries', 'Total.Minor.Injuries', 
                   'Total.Uninjured', 'fatality_level', 'sorted_damage']

# Set plot size and layout
fig, axes = plt.subplots(nrows=len(columns_to_plot), figsize=(12, 5 * len(columns_to_plot)))

# Loop through each column to create a boxplot
for i, column in enumerate(columns_to_plot):
    sns.violinplot(
        data=new_df,
        x='Broad.phase.of.flight',
        y=column,
        ax=axes[i],
        palette='Set3'
    )
    axes[i].set_title(f'{column} Across Broad Phases of Flight')
    axes[i].set_xlabel('Broad Phase of Flight')
    axes[i].set_ylabel(column)
    axes[i].tick_params(axis='x', rotation=45)

# Adjust spacing
plt.tight_layout()
plt.show()
plt.show()
```

Q10. Injury Analysis Across Locations: How does the total number of injuries (fatal, serious, minor) vary across different countries or locations? Create visualizations such as maps or bar plots to show injury distribution by location.

```python
top_n=10
injury_levels = ['fatality_level', 'Total.Serious.Injuries', 'Total.Minor.Injuries']

# Create subplots (3 rows and 1 column)
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 18))

# Loop over injury levels and plot each one in a separate subplot
for i, injury in enumerate(injury_levels):
    # Sort the data by the current injury level in descending order and select the top N
    top_data = new_df.sort_values(by=injury, ascending=False).head(top_n)
    
    # Plot the data
    sns.barplot(x='Country', y=injury, data=top_data, ax=axes[i])
    axes[i].set_title(f'Top {top_n} Countries by {injury.replace("_", " ").title()}')
    axes[i].set_xlabel('Country')
    axes[i].set_ylabel(f'{injury.replace("_", " ").title()}')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability

# Adjust layout to avoid overlapping subplots
plt.tight_layout()
plt.show()
```
